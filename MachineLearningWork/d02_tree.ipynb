```python
# 무게  표면상태     결과
# 270  울퉁불퉁(0)  오렌지(0)
# 260  울퉁불퉁(0)  오렌지(0)
# 250  울퉁불퉁(0)  오렌지(0)
# 220    매끈(1)    사과(1)
# 230    매끈(1)    사과(1)
# 240    매끈(1)    사과(1)

# 250    매끈(1)
```

```python
# DecisionTreeclassifier 사용하여 250 매끈(1) 사과인지 오렌지인지 예측
from sklearn import tree
input_data = [[270,0],[260,0],[250,0],[220,1],[230,1],[240,1]]
fruits = [0,0,0,1,1,1]
clf = tree.DecisionTreeClassifier()
clf = clf.fit(input_data, fruits)
clf
```

```python
print(clf.predict([[250,1]]))
```

```python
# 시각화
import matplotlib.pyplot as plt
plt.figure(figsize=(20,17))
tree.plot_tree(clf,
               impurity=True,
               filled=True,
               rounded=True
               )
```

```python
print(clf.score(input_data, fruits))
```
