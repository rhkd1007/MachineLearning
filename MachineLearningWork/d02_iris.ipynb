sklearn.datasets 내의 모듈은 사이킷런에서 자체적으로 제공하는 데이터 세트를 생성하는 모듈의 모임이다.

sklearn.tree 내의 모듈은 트리 기만 ML 알고리음(Machine Learning Algorithm)을 구현한 클래스의 모임

sklearn.model_selection 학습, 검증, 예측데이터로 분리하거나 최적의 하이퍼파라미트로 평가하기 위한 모듈

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import pandas as pd

# iris.data : Iris 데이터 세트에서 피처(feature)만으로 된 데이터를 numpy로 가지고 있다.
# iris.target : 붓꽃 데이터 세트에서 레이블(결정 값) 데이터를 numpy로 가지고 있다.

iris = load_iris()
iris_data = iris.data
iris_label = iris.target
print('iris.data : ', iris_data)
print('iris.target : ', iris_label)
print('iris.target명 : ', iris.target_names)
```

```python
# 붓꽃 데이터 세트를 DataFrame으로 변환
# iris.target value 값으로 한 label 속성 추가

iris_df = pd.DataFrame(data = iris_data, columns=iris.feature_names)
iris_df

iris_df['label'] = iris.target
iris_df
```

random_state를 지정하지 않으면 학습/테스트 수행할때마다 임의의 데이터를 만듦으로 임의의 수 부여(어떤 숫자여도 상관없음)

test_size=0.2 는 전체 데이터 세트 중 테스트 데이터 세트의 비율 디폴트값은 0.25

```python
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, random_state=11, test_size=0.2)
print(iris_df.info)
X_train.shape
```

```python
#훈련
dt_clf = DecisionTreeClassifier(random_state=11)
dt_clf.fit(X_train, y_train)
```

```python
#예측
pred = dt_clf.predict(X_test)
#accuracy_score()의 첫 번째 파라미터로 실제 레이블 데이터세트,
# 두 번째 파라미터로 예측 레이블 데이터 세트를 입력하면 된다.

from sklearn.metrics import accuracy_score
print('accuracy : ', round(accuracy_score(y_test, pred),4))
```

```python
#시각화
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(20,17))
feature_names = ['sepal length','sepal width','petal length','petal width']
tree.plot_tree(dt_clf,
               feature_names=feature_names,
               impurity=True,
               filled=True,
               rounded=True
               )
```

```python
dt_clf.feature_importances_
```

```python
for name, value in zip(feature_names, dt_clf.feature_importances_):
    print('{0}:{1:.3f}'.format(name,value))
```

```python
import seaborn as sns
sns.barplot(x=dt_clf.feature_importances_, y=feature_names)
```

```python
# 앞의 예제와 다르게 테스트 데이터 세트를 전체의 30%로, random_state=121로 설정
dt_clf2 = DecisionTreeClassifier()
iris_data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris_data.data, iris_data.target, random_state=121, test_size=0.3)
dt_clf2.fit(X_train,y_train)
pred = dt_clf2.predict(X_test)
print('예측정확도 : ', round(accuracy_score(y_test,pred),4))
```
