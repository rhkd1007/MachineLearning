```python
# 더이상 필요 없는 열 삭제 : 'name','ticket','Sibsp','parch','passengerid'
test = df.drop(['Name','Ticket','SibSp','Parch','PassengerId'],axis=1)
test
```

```python
# 더이상 필요 없는 열 삭제 : 'name','ticket','Sibsp','parch','passengerid'
target_data = df_test.drop(['Name','Ticket','SibSp','Parch','PassengerId'],axis=1)
target_data
```

Training set : 모델의 학습에 사용하는 데이터

Validation set : 모델 제작 과정 중, 학습된 모델의 성능을 측정하기 위한 데이터

Test set : 모델의 최종 성능을 평가하기 위한 데이터


```python
train_data = test.drop('Survived', axis = 1)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_valid, y_train, y_vaild = train_test_split(train_data, target_data)
print(train_data.shape, x_train.shape, x_valid.shape)
```

```python
# sklearn_learn의 Decision Tree Classifier를 사용하여 학습하기
from sklearn.tree import DecisionTreeClassifier

# Decison Tree로 모델 생성
tree_dec = DecisionTreeClassifier()

#학습
tree_dec.fit(x_train,y_train)
```

```python
tree_dec.score(x_train,y_train)
```

```python
tree_dec.score(x_valid,y_vaild)
```

```python
# 학습된 모델로 x_train의 결과를 얻어낸 다음
# 그 예측과 y_train을 비교해서 정확도를 보임
print('training set acccuracy : ', tree_dec.score(x_train,y_train))
print('Validation set acccuracy : ', tree_dec.score(x_valid,y_vaild))
```

```python
# 만들어진 Decision Tree로 test 데이터 판단
prediction = tree_dec.predict(test)
prediction
```

```python
submit = pd.DataFrame({
    'PassengerId':df_test['PassengerId'],
    'Survived':prediction
})
submit.to_csv('submit.csv', index=False)
```
