# 보스톤 주택가격 데이터에 머신러닝 기반의 회귀 분석 수행

```python
from sklearn.datasets import load_boston
import pandas as pd
boston = load_boston()
boston.DESCR
```

```python
boston_df = pd.DataFrame(boston.data, columns=boston.feature_names)
boston_df.head()
```

CRIM : 지역별 범죄 발생률
ZN : 25,000평방피트를 초과하는 거주 지역
INDUS : 비상업 지역의 넓이 비율
CHAS : 찰스강의 더미변수(1은 강의 경계, 0 은 경계 아님)
NOX : 일산화 질소 농도
RM :거주할 수 있는 방 개수
AGE : 1940년 이전에 건축된 주택 비율
DIS : 5개 주요 고용센터까지 가중거리
RAD : 고속도로 접근 용이도
TAX : 10,000 달러당 재산세 비율
PTRATIO : 지역의 교사와 학생 수 비율
B : 지역의 흑인 거주 비율
LSTAT : 하위 계층 비율
Price(MEDV) : 본인 소유 주택 가격의 중앙값

```python
boston.target
```

```python
# boston_df 에 PRICE 라는 컬럼 추가 value 는 boston.target
boston_df['PRICE'] = boston.target
boston_df.head()
```

```python
print('보스턴 주택 가격 데이터 셋 크기 :', boston_df.shape)
```

```python
boston_df.info()
```

```python
boston_df.isnull().sum()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# x, y (data 와 label) 분할하기
Y = boston_df['PRICE']
X = boston_df.drop(['PRICE'],axis=1,inplace=False)
X
```

```python
Y
```

```python
#훈련용 데이터와 평가용 
X_train, X_test, Y_train, Y_test =  train_test_split(X,Y, test_size=0.3,
                                                     random_state=156)
```

```python
# 선형회귀 분석 모델 생성
lr = LinearRegression()
# 훈련
lr.fit(X_train, Y_train)
```

```python
# 예측
Y_predict = lr.predict(X_test)
Y_predict
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
```

```python
mse = mean_squared_error(Y_test, Y_predict)
rmse = np.sqrt(mse)
```

```python
print('mse : ', mse, 'rmse : ', rmse)
```

```python
print('Y절편 값 : ', lr.intercept_)
print('회귀계수(기울기) : ',np.round(lr.coef_, 1))
```

```python
coef = pd.Series(data =np.round(lr.coef_,2), index = X.columns)
coef.sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x_features = ['CRIM','ZN','INDUS','CHAS','NOX','RM',
              'AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT'] # 한줄에 3개씩
fig, axs = plt.subplots(figsize=(16,16), ncols=3, nrows=5)

for i, feature in enumerate(x_features):
    row = int(i/3)
    col = i%3
    # regplot : 산점도와 선형 회귀 그래프를 함께 그려줌
    sns.regplot(x=feature, y='PRICE', data=boston_df, ax=axs[row][col])
```
