```python
import pandas as pd

# train.csv 읽어 5줄 출력
df = pd.read_csv('train.csv')
df.head()
```

```python
# Pclass Sex Embarked 만 추출해서 5개 출력
df[['Pclass', 'Sex', 'Embarked']].head()
```

```python
# test 데이터 불러오기 5줄 출력
df_test=pd.read_csv('test.csv')
df_test.head()
```

```python
df['Age']
```

```python
# null값 보는 법
df_test.isnull().sum()
```

```python
# Age : 비어있는 값을 나이의 평균으로 채우기
df['Age'] = df['Age'].fillna(df['Age'].mean())
df.head(10)
```

```python
df.isnull().sum()
```

```python
# df_test에서 Age에 null 값을 평균값으로 대체
df_test['Age'] = df_test['Age'].fillna(df_test['Age'].mean())
df_test.head(10)
```

```python
df_test.isnull().sum()
```

```python
# Fare 빈칸(NaN)을 평균으로 대체 (훈련용, 테스트용 모두 대체)
df_test['Fare'] = df_test['Fare'].fillna(df_test['Fare'].mean())
df['Fare'] = df['Fare'].fillna(df['Fare'].mean())
```

```python
# df.head(10)
# df_test.head(10)
df_test.isnull().sum()
```

```python
# SibSp와 Parch를 더하여 FamilySize 생성 (훈련용, 테스트용 모두 생성)
df['FamilySize'] = df['SibSp'] + df['Parch']
df_test['FamilySize'] = df_test['SibSp']+df_test['Parch']
```

```python
# df
df_test
```

```python
df['Cabin']
```

```python
# Cabin(객실) 정보에서 번호만 삭제하고 객실 종류에 해당하는 알파벳만 남기기
df['Cabin'].str[0]
```

```python
df['Cabin'] = df['Cabin'].str[0]
df['Cabin']
```

```python
# Cabin의 객실 NaN : 가장 많은 알파벳으로 대체
df['Cabin'].value_counts()
```

```python
df['Cabin'] = df['Cabin'].fillna('C')
df['Cabin']
```

```python
# Cabin test 셋에도 적용
df_test['Cabin'] = df_test['Cabin'].str[0]
df_test['Cabin'].value_counts()
df_test['Cabin'] = df_test['Cabin'].fillna('C')
df_test['Cabin']
```

```python
df['Embarked'].value_counts()
```

```python
df['Embarked'] = df['Embarked'].fillna('S')
df.isnull().sum()
```

```python
# df_test['Embarked'].value_counts()
df_test['Embarked'] = df_test['Embarked'].fillna('S')

df_test.isnull().sum()
```

```python
df['Cabin']
```

```python
# 문자를 숫자로 변환
# df.loc[df['Cabin']=='C', 'Cabin']=0
# df.loc[df['Cabin']=='B', 'Cabin']=1

df['Cabin'] = df['Cabin'].map({
    'C' : 0,
    'B' : 1,
    'D' : 2,
    'E' : 3,
    'A' : 4,
    'F' : 5,
    'G' : 6,
    'T' : 7
})
```

```python
df['Cabin'].value_counts()
```

```python
# test 셋 적용
df_test['Cabin'] = df_test['Cabin'].map({
    'C' : 0,
    'B' : 1,
    'D' : 2,
    'E' : 3,
    'A' : 4,
    'F' : 5,
    'G' : 6,
    'T' : 7
})

df_test['Cabin'].value_counts()
```

```python
# Embarked의 문자값을 숫자로 변환
df['Embarked'] = df['Embarked'].map({
    'S' : 0,
    'C' : 1,
    'Q' : 2
})
df['Embarked'].value_counts()
```

```python
df_test['Embarked'] = df_test['Embarked'].map({
    'S' : 0,
    'C' : 1,
    'Q' : 2
})
df_test['Embarked'].value_counts()
```

```python
# Sex의 문자값을 숫자로 변환
df['Sex'] = df['Sex'].map({
    'male' : 0,
    'female' : 1,
})
df['Sex'].value_counts()
```

```python
df_test['Sex'] = df_test['Sex'].map({
    'male' : 0,
    'female' : 1,
})
df_test['Sex'].value_counts()
```

```python
# 필요없는 열 삭제
train = df.drop(['Name', 'Ticket', 'SibSp', 'Parch', 'PassengerId'], axis=1)
# train.head()
test = df_test.drop(['Name', 'Ticket', 'SibSp', 'Parch', 'PassengerId'], axis=1)
# test.head()
```

```python
train_data = train.drop(['Survived'], axis=1)   # 학습용 데이터
target_data = train['Survived'] #학습용 데이터 label
train_data.head()
# test_data = test.drop(['Survived'], axis=1)   # 학습용 데이터
# test_data = train['Survived'] #학습용 데이터 label
# test_data.head()
```

```python
target_data.head()
```

Training set : 모델의 학습에 사용하는 데이터

Validation set : 모델 제작 과정 중, 학습된 모델의 성능을 측정하기 위한 데이터

Test set : 모델의 최종 성능을 평가하기 위한 데이터


```python
from sklearn.model_selection import train_test_split

x_train, x_valid, y_train, y_valid = train_test_split(train_data, target_data, test_size=0.3)
print(train_data.shape, x_train.shape, x_valid.shape)
```

```python
# sklearn-learn의 Decision Tree Classifier를 사용하여 학습하기
from sklearn.tree import DecisionTreeClassifier

# Decision Tree로 모델 생성
tree_dec = DecisionTreeClassifier()

# 학습
tree_dec.fit(x_train, y_train)
```

```python
tree_dec.score(x_train, y_train)
```

```python
tree_dec.score(x_valid, y_valid)
```

```python
# 학습된 모델로 x_train의 결과를 얻어낸 다음
# 그 예측과 y_train을 비교해서 정확도를 보임
print('trainning set accuracy : ', tree_dec.score(x_train, y_train))
print('Validation set accuracy : ', tree_dec.score(x_valid, y_valid))
```

```python
# 만들어진 Decision Tree로 test 데이터 판단
prediction = tree_dec.predict(test)
prediction
```

```python
submin = pd.DataFrame({
    'PassengerId' : df_test['PassengerId'],
    'Survived' : prediction
})
submin.to_csv('submit.csv', index=False)
```

```python
# 시각화
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(20,17))
feature_names=['Pclass', 'Sex', 'Age', 'Fare', 'Cabin', 'Embarked', 'FamilySize']
tree.plot_tree(tree_dec,
               class_names=['Unsurvived', 'Survived'],
               feature_names=feature_names,
               impurity=True,
               filled=True,
               rounded=True)
```

```python
# Pclass Sex Age Fare Cabin Embarked FamilySize
print("importance : ", tree_dec.feature_importances_)
```

```python
for name, value in zip(feature_names,tree_dec.feature_importances_):
    print('{0}:{1:.3f}'.format(name,value))
```

```python
import seaborn as sns
sns.barplot(x= tree_dec.feature_importances_, y= feature_names)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
parameters = {
    'max_depth' : [2,3,5,10],
    'min_samples_split' : [2,3,5],
    'min_samples_leaf' : [1,5,8]
}
# parameters = {
#     'max_depth' : [8,10,12,15],
#     'min_samples_split' : [4,5,7],
#     'min_samples_leaf' : [1,2,3]
# }
# score : 예측 성능을 측정할 평가 방법 보통은 정확도로 성능 평가를 한다.
# cv : 교차 검증에서 몇개로 분할되는지 지정한다.
grid_dclf = GridSearchCV(tree_dec, 
                         param_grid=parameters,
                         scoring='accuracy',
                         cv=5)
grid_dclf
```

```python
grid_dclf.fit(x_train, y_train)
```

```python
print(f'GridSearchCV 최적 하이퍼 파라미터:{grid_dclf.best_params_}') 
```

```python
print(f'GridSearchCV 최고 정확도:{round(grid_dclf.best_score_,4)}')
```

```python
# GridSearchCV의 최적 하이퍼 파라미터로 학습된 Estimator로 예측 및 평가 수행
best_dclf = grid_dclf.best_estimator_
best_predictions = best_dclf.predict(x_valid)
accuracy = accuracy_score(y_valid, best_predictions)
print(f'테스트 세트에서의 DecisionTreeClassifier 정확도 : {round(accuracy,4)}')
```

```python
# DecisionTreeClassifier 학습/예측/평가
tree_dec.fit(x_train, y_train)
dt_pred = tree_dec.predict(x_valid)
print('DecisionTreeClassifier 정확도 : {0:.4f}' .format(accuracy_score(y_valid, dt_pred)))
```

```python
# DecisionTreeClassifier 검증용 예측/평가
vprediction = tree_dec.predict(x_valid)
accuracy = accuracy_score(vprediction, y_valid)
print('accuracy : ',accuracy)

length = y_valid.shape[0]
print('length : ',length)
print(f'총 {length}명 중 {accuracy * 100:.3f}% 정확도로 생존을 맞춤')
```

```python
# 시각화 ==> best tree
plt.figure(figsize=(20,17))
feature_names=['Pclass', 'Sex', 'Age', 'Fare', 'Cabin', 'Embarked', 'FamilySize']
tree.plot_tree(best_dclf,
               class_names=['Unsurvived', 'Survived'],
               feature_names=feature_names,
               impurity=True,
               filled=True,
               rounded=True)
```
