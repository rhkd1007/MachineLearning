```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(wx):
    return 1/(1+np.exp(-wx))

wx = np.arange(-8.0, 8.0, 0.1)
y = sigmoid(wx)

plt.plot(wx,y,'b')
plt.grid()
plt.xlabel('z', size=10)
plt.ylabel('(z)', size=10)
plt.title('Logistic | sigmoid function')
plt.show()
```

승산(Odds) : 

확률에서 시작되어 y=1 일 성공이라 가정할때 실패(1-p)에 대한 성공(p)할 확률 비를 의미

odd(p) = p/(1-p)

예) 파란구슬 4개 빨강구슬 6개일때 파란구슬을 선택할 확률 4/10 ==> 40%

odds) 파란구슬(4)/빨간구슬(6) 즉, 파란구슬(4)/1-파란구슬(4)

예) 대비 몇 % : 파(5),빨(5) : 5/5 = 1 (100%) /// 파(7),빨(3) :  7/3 = 2.333(약 233%)



odds의 문제점 p 가 1에 가까울 수록 승산이 치솟음



로짓변환 그래프

입력 0~1사이에 대하여 출력갑은 -무한대 ~ 무한대를 가짐

0<=p<=1 가 만족



```python
# 승산(Odds) : 
# 확률에서 시작되어 y=1 일 성공이라 가정할때 실패(1-p)에 대한 성공(p)할 확률 비를 의미
# odd(p) = p/(1-p)
# 예) 파란구슬 4개 빨강구슬 6개일때 파란구슬을 선택할 확률 4/10 ==> 40%
# odds) 파란구슬(4)/빨간구슬(6) 즉, 파란구슬(4)/1-파란구슬(4)
# 예) 대비 몇 % : 파(5),빨(5) : 5/5 = 1 (100%) /// 파(7),빨(3) :  7/3 = 2.333(약 233%)

### odds의 문제점 p 가 1에 가까울 수록 승산이 치솟음

### 로짓변환 그래프
# 입력 0~1사이에 대하여 출력갑은 -무한대 ~ 무한대를 가짐
# 0<=p<=1 가 만족
```

```python
def odds(p):
    return p/(1-p)

# 확률 p : 0~1
p = np.arange(0.001,0.999,0.001)
z = [odds(n) for n in p]

plt.plot(p,z)
plt.ylim([0,200.0])
plt.xlabel("p")
plt.ylabel("odds")

plt.show()
```

```python
# odds > logit > sigmoid

def logit1(p):
    return np.log(p/(1-p))

p = np.arange(0.001,0.999,0.001)
z = [logit1(n) for n in p]

plt.plot(p,z)
plt.xlabel("p")
plt.ylabel("logit")

plt.show()
```
